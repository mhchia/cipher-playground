```python
n = 12
target_orders = [1] * 1 + [2]*2 + [3]*9
```

```python
# Obvious not C12 since it has order-12 element
C12 = CyclicPermutationGroup(n); C12
```

```python
C6 = CyclicPermutationGroup(6)
C2 = CyclicPermutationGroup(2)
# help(C6.direct_product)
C6xC2,iota1,iota2,pr1,pr2 = C6.direct_product(C2); assert C6xC2.order() == n; C6xC2
```

```python
sorted([x.order() for x in C6xC2])
```

```python
D6 = DihedralGroup(6); assert D6.order() == n; D6
```

```python
sorted([x.order() for x in D6])
```

```python
A4 = AlternatingGroup(4); assert A4.order() == n; A4
```

```python
[x.order() for x in A4]
```

```python
Dic6 = DiCyclicGroup(3); assert Dic6.order() == n, f"{Dic6.order()=}, {n=}"; Dic6
```

```python
sorted([x.order() for x in Dic6])
```
