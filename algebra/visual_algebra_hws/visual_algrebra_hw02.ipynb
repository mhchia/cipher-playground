```python
# Use CyclotomicField to construct the matrices. Can use finite fields as well
K = CyclotomicField(4); K
```

```python
i = K.gen(); i
```

```python
A = Matrix(K, [[i, 0], [0, -i]]); A
```

```python
B = Matrix(K, [[0, -1], [1, 0]]); B
```

```python
A^2 == -1, B^2 == -1, A^4 == 1, B^4 == 1
```

```python
G = MatrixGroup([A, B]); G
```

```python
G.order()
```

```python
elements = G.list(); elements
```

```python
# the names of the elements
names = ['I', '-I', "-A", "A", "-B", "B", "AB", "-AB"]
```

```python
G.cayley_table(names=names, elements=elements)
```

```python
print("Orders: ")
list(zip(names, map(lambda x: x.order(), elements)))
```

```python
G.is_isomorphic(Q8)
```
